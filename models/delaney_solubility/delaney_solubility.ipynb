# Delaney Solubility

# Load dataset

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

# Data Preparation

## Data separation X and y


```python
y = df['logS']
y
```

```python
X = df.drop('logS', axis=1)
X
```

## Data splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

# Model Building

## Linear Regression

### Training the model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

### Applying the model to make a prediction

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)

print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

### Evaluate Model Performence

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

## Random Forest

### Training the model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

### Applying the model to make a prediction

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

### Evaluate Model Performence

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

## Support Vector Regression (SVR)

### Training the model

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV

pipeline = make_pipeline(StandardScaler(), SVR())

param_grid = {
    'svr__C': [0.1, 1, 10, 100],
    'svr__epsilon': [0.01, 0.1, 0.2],
    'svr__gamma': ['scale', 'auto'],
    'svr__kernel': ['rbf']  # you can add 'linear' if you want to test it too
}

grid_search = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train, y_train)

# The best estimator can be accessed via grid_search.best_estimator_
# The best parameters can be accessed via grid_search.best_params_
```

### Applying the model to make a prediction

```python
y_svr_train_pred = grid_search.best_estimator_.predict(X_train)
y_svr_test_pred = grid_search.best_estimator_.predict(X_test)
```

### Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score

svr_train_mse = mean_squared_error(y_train, y_svr_train_pred)
svr_train_r2 = r2_score(y_train, y_svr_train_pred)

svr_test_mse = mean_squared_error(y_test, y_svr_test_pred)
svr_test_r2 = r2_score(y_test, y_svr_test_pred)
```

```python
svr_results = pd.DataFrame(['Support Vector Regression', svr_train_mse, svr_train_r2, svr_test_mse, svr_test_r2]).transpose()
svr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
svr_results
```

## Model Comparison

```python
df_models = pd.concat([lr_results, rf_results, svr_results], axis=0).reset_index(drop=True)
df_models
```

```python
df_models.reset_index(drop=True)
```

🧠 Interpretation
* SVR now outperforms both other models significantly, especially on test data.

* Train MSE is lower than the others, but not absurdly small, so overfitting is not a big concern.

* Test R² of 0.84 is excellent — it means 84% of the variance in solubility is explained by the model.
